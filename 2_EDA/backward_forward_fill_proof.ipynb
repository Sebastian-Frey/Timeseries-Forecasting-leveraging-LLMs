```python
import pandas as pd
from pathlib import Path

# 1) How many keywords have missing at the last pre-test week?
#    Suppose last 6 weeks are test; find week T-6 (the last train week)
df = pd.read_parquet("/content/drive/MyDrive/colab_data/cleaned_cpu/weekly_aggregated_by_week_keyword.parquet")
# get last week per keyword
df["year"], df["week_num"] = zip(*df["week"].str.split("-").map(lambda x: (int(x[1]), int(x[0]))))
df = df.sort_values(["keyword","year","week_num"])

last6 = df.groupby("keyword").tail(1)
last_train = df.groupby("keyword").nth(-2)  # the last week before the 6-week holdout

n_boundary_missing = last_train["cpc_week"].isna().sum()
print("Keywords with missing cpc at last train week:", n_boundary_missing)

# 2) Estimate how widespread any two-sided fill influence is:
# Count keywords with ANY NaN during the entire series (pre-interpolation)
has_any_missing = df.groupby("keyword")["cpc_week"].apply(lambda s: s.isna().any()).sum()
print("Keywords that had any missing cpc_week:", has_any_missing)
```
